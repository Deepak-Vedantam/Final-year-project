# Preprocessing

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
df=pd.read_csv("../input/final-year-project/Final Combined.csv")
df.head(5)
```

```python
df.dropna()
```

```python
X=list(df['Text'].astype("str"))
X[:5]
```

```python
sen_len=[]
for i in X:
    sen_len.append(len(i))
print(len(sen_len))
```

```python
df_len=pd.DataFrame(sen_len)
df_len.describe()
```

```python
plt.subplots(figsize=(20,4))
plt.title("Length Variattion of Sentences")
sns.scatterplot(x=["MIN","MEAN","MAX"],y=[df_len.min(axis=0)[0],df_len.mean(axis=0)[0],df_len.max(axis=0)[0]])
plt.show()
```

```python
st_final=""
for i in tqdm.tqdm(X):
    st_final+=(i[1:len(i)-1])
```

```python
filters_custom=set(st_final)
```

```python
sent=st_final.split(" ")
all_words=len(sent)
```

```python
dict_len=len(set(sent))
print(dict_len,all_words)
```

```python
st_filter=""
for i in filters_custom:
    st_filter+=i
st_filter+="!#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n"
print(st_filter)
```

```python
for i in st_filter:
    if i.lower() in "0123456789abcdefghijklmnopqrstuvwxyz":
        st_filter=st_filter.replace(i,'')
print(st_filter)
        
```

```python
tokenizer=Tokenizer(num_words=dict_len,filters=st_filter,lower=True,oov_token="UNK")
tokenizer.fit_on_texts(X)
```

```python
len(tokenizer.word_index)
```

```python
X_train=tokenizer.texts_to_sequences(X)
X_train_padded=pad_sequences(X_train,maxlen=138,padding='post',truncating='post')
```

```python
X_train_padded[0]
```

```python
Y_train=df['oh_label'].astype('float')
```

```python
type(Y_train[0])
```

```python
set(Y_train)
```

# The Parameters

```python
word_dict_len=len(tokenizer.word_index)
embed_dim=128
in_len=138
```

# The Model

```python
import tensorflow
from tensorflow.keras.models import Sequential 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding,Bidirectional,LSTM,Dense,Dropout
from tensorflow.keras.utils import to_categorical
```

```python
model=Sequential()
model.add(Embedding(word_dict_len,embed_dim,input_length=in_len))
model.add(Dropout(0.4))
model.add(Bidirectional(LSTM(in_len,return_sequences=True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(in_len*2,return_sequences=True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(in_len*4)))
model.add(Dense(in_len*4,activation="relu"))
model.add(Dense(in_len*2,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(in_len,activation='relu'))
model.add(Dense(in_len//2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
print(model.summary())
```

```python
model.compile(optimizer='rmsprop',metrics=['accuracy'],loss='binary_crossentropy')
```

```python
hist=model.fit(X_train_padded,Y_train,epochs=5,validation_split=0.2)
```

```python
hist=model.fit(X_train_padded,Y_train,epochs=2,validation_split=0.2)
```

```python
plt.plot(hist.history['accuracy'],c='b',label='Training')
plt.plot(hist.history['val_accuracy'],c='r',label='Validation')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(hist.history['loss'],c='b',label='Training')
plt.plot(hist.history['val_loss'],c='r',label='Validation')
plt.legend(loc='upper right')
plt.show()
```
